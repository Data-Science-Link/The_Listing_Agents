```python
import pandas as pd
import numpy as np
import seaborn as sns
# homes = pd.read_csv('homes.csv')
```

```python
from read_path_module import read_data_relative_path

homes = read_data_relative_path(relative_dataset_path = './data/kaggle/train.csv',
                        data_type='csv'
                       )

homes.head(3)
```

# LotFrontage: (0, 1, 2) vs. else


```python
homes.LotFrontage.agg(['max','min'])
homes['LotFrontage_bin']= pd.cut(homes['LotFrontage'],15, labels = False)
homes.groupby('LotFrontage_bin').agg(['count','mean','max','min'])['SalePrice']
```

```python
cat = 'LotFrontage_bin'

homesGrid = sns.FacetGrid(homes, row= cat, hue= cat, palette="Set1", height  = 3, aspect = 4 )

homesGrid.map(sns.distplot,'saleprice_bin', kde = False)
```

# LotArea: Discussion


```python
homes.LotArea.agg(['max','min'])
homes['LotArea_bin']= pd.cut(homes['LotArea'],100, labels = False)
homes.groupby('LotArea_bin').agg(['count','mean','std','max','min'])['SalePrice']
```
